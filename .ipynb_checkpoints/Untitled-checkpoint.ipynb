```python
import numpy as np
import pandas as pd
import glob
import cv2
import os
import torch
import matplotlib.pyplot as plt
import torchvision
```

## Import dataset

```python
train_data=pd.read_csv('./sign_mnist_train.csv').fillna(0)

train_data.head()
```

```python
test_data=pd.read_csv('./sign_mnist_test.csv').fillna(0)
test_data.head()
```

## Extracting the labels and Images

```python
y_train=train_data['label']
y_test=test_data['label']
print(y_train.shape)
print(y_test.shape)
```

```python
## dropping labels from x_train/test
x_train=train_data.drop(['label'],axis=1)
x_train.head()
```

```python
x_test=test_data.drop(['label'],axis=1)
x_test.head()
```

```python
#converting to numpy arrays

x_train=x_train.to_numpy()
x_test=x_test.to_numpy()

print(x_train.shape)
print(x_test.shape)
```

## Displaying Images




```python
X_train=np.reshape(x_train,(-1,28,28))
print(X_train.shape)
```

```python
X_test=np.reshape(x_test,(-1,28,28))
print(X_test.shape)
```

```python
#plotting image
for i in range(0,5):
    plt.imshow(X_train[i,:,:],cmap='gray')
    plt.show()
```
